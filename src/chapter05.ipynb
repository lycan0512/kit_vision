```python
import cv2

src = cv2.imread("tomato.jpg")
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(hsv)
h_red = cv2.inRange(h, 0, 5)

dst = cv2.bitwise_and(hsv, hsv, mask = h_red)
dst = cv2.cvtColor(dst, cv2.COLOR_HSV2BGR)

cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread("tomato.jpg")
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(hsv)

orange = cv2.inRange(hsv, (8, 100, 100), (20, 255, 255))
blue = cv2.inRange(hsv, (110, 100, 100), (130, 255, 255))
mix_color = cv2.addWeighted(orange, 1.0, blue, 1.0, 0.0)

dst = cv2.bitwise_and(hsv, hsv, mask = mix_color)
dst = cv2.cvtColor(dst, cv2.COLOR_HSV2BGR)

cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread("swan.jpg")
_, binary = cv2.threshold(src, 127, 255, cv2.THRESH_BINARY)

cv2.imshow("binary", binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread("swan.jpg")
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 33, -5)

cv2.imshow("binary", binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 연산
import cv2

src = cv2.imread("swan.jpg")

offset = np.full(src.shape, 50, dtype =np.uint8)

operated = cv2.add(src, offset)

dst = np.concatenate((src, operated), axis=0)

cv2.namedWindow("dst", flags=cv2.WINDOW_FREERATIO)
cv2.resizeWindow("dst", 640, 960)
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 연산
import cv2

src = cv2.imread("swan.jpg")

offset = np.full(src.shape, 50, dtype =np.uint8)

operated = cv2.subtract(src, offset)

dst = np.concatenate((src, operated), axis=0)

cv2.namedWindow("dst", flags=cv2.WINDOW_FREERATIO)
cv2.resizeWindow("dst", 640, 960)
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 원본 이미지 히스토그램
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("swan.jpg")

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    plt.ylim([0, 42000])
plt.show()
```
